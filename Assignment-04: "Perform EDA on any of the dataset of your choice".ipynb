# **EDA**


# **1. Importing the required libraries for EDA**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# **2. Loading the data into the data frame**

```python
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
df.head()
```

```python
df.tail(15)
```

# **3. Checking the types of data**

```python
df.dtypes
```

# **4. Dropping irrelevant columns**

```python
pd.unique(df['housing_median_age'])
```

```python
df = df.drop(['housing_median_age'], axis=1)
df.head(5)
```

# **5. Renaming the columns**

```python
df = df.rename(columns={"housing_median_age": "age", "total_rooms": "rooms", "total_bedrooms": "bedrooms", "population": "pop","households": "house", "median_income": "income", "median_house_value": "value" })
df.head(5)
```

# **6. Dropping the duplicate rows**

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

# **7. Dropping the missing or null values.**

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

# **8. Detecting Outliers**

```python
sns.boxplot(x=df['latitude'])
```

```python
sns.boxplot(x=df['value'])
```

# **9. Plot different features against one another**

**Histogram**

```python
df.longitude.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of bedrooms by longitude")
plt.ylabel('Number of bedrooms')
plt.xlabel('Longitude');
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['total_rooms'])
```

**Scatter Plot**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['total_rooms'], y=df['median_house_value'])
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['total_rooms'],y=df['median_house_value'])
```

**Heat Maps**

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
```
