# **Q.10**

```python
import youtube_dl
import os
from moviepy.editor import VideoFileClip
from concurrent.futures import ThreadPoolExecutor

# Set up the options for downloading videos
ydl_opts = {
    'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',
    'ignoreerrors': True,
    'no_warnings': True,
    'postprocessors': [{
        'key': 'FFmpegVideoConvertor',
        'preferedformat': 'mp4',
    }],
    'outtmpl': 'video_%(autonumber)s.mp4',
    'max_downloads': 100
}

# Set up the thread pool executor
executor = ThreadPoolExecutor(max_workers=5)

# Download and convert video to audio
def download_and_convert(url):
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(url, download=True)
        video_path = ydl.prepare_filename(info_dict)
        audio_path = os.path.splitext(video_path)[0] + ".mp3"

        video = VideoFileClip(video_path)
        audio = video.audio
        audio.write_audiofile(audio_path)

        print(f"Converted {video_path} to audio.")

# List of YouTube video URLs to download
video_urls = [
    'https://www.youtube.com/watch?v=VIDEO1',
    'https://www.youtube.com/watch?v=VIDEO2',
    # Add more URLs here...
]

# Download and convert each video in parallel
for url in video_urls:
    executor.submit(download_and_convert, url)

executor.shutdown(wait=True)

print("Conversion of all videos to audio complete!")
```

# **Q.11**

```python
import os
import requests
from PIL import Image
from concurrent.futures import ThreadPoolExecutor

# Set up the search query and the number of images to download
search_query = "dog"
num_images = 500

# Set up the thread pool executor
executor = ThreadPoolExecutor(max_workers=10)

# Download and resize the images
def download_and_resize(url):
    response = requests.get(url, stream=True)
    image_path = os.path.join("images", url.split("/")[-1])

    with open(image_path, 'wb') as image_file:
        for chunk in response.iter_content(1024):
            image_file.write(chunk)

    image = Image.open(image_path)
    resized_image = image.resize((int(image.width/2), int(image.height/2)))
    resized_image.save(image_path)

    print(f"Resized {image_path} to 50%.")

# Set up the Google Images search URL
search_url = f"https://www.google.com/search?q={search_query}&tbm=isch"

# Perform the search and extract image URLs
response = requests.get(search_url)
image_urls = response.text.split('src="')[1:num_images+1]

# Download and resize each image in parallel
for url in image_urls:
    url = url.split('"')[0]
    executor.submit(download_and_resize, url)

executor.shutdown(wait=True)

print("Resizing of all images complete!")
```

# **Q.13 (i) to (ii)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage


np.random.seed(42)
dataset = pd.DataFrame(np.random.uniform(-10, 10, size=(500, 4)), columns=['Column1', 'Column2', 'Column3', 'Column4'])
dataset[['Column5', 'Column6', 'Column7', 'Column8']] = np.random.uniform(10, 20, size=(500, 4))
dataset[['Column9', 'Column10']] = np.random.uniform(-100, 100, size=(500, 2))


scaler = StandardScaler()
scaled_dataset = scaler.fit_transform(dataset)


inertia_values = []
silhouette_scores = []
max_clusters = 10

for n_clusters in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(scaled_dataset)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(kmeans.score(scaled_dataset))


plt.figure(figsize=(10, 6))
plt.plot(range(2, max_clusters + 1), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve for K-Means Clustering')
plt.show()


plt.figure(figsize=(10, 6))
plt.plot(range(2, max_clusters + 1), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-Means Clustering')
plt.show()


dendrogram_data = scaled_dataset[:100]


linkage_matrix = linkage(dendrogram_data, method='ward')


plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix, truncate_mode='level', p=3)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.title('Dendrogram for Hierarchical Clustering')
plt.show()
```

# **Q.15 (i) to (iii)**

```python
import numpy as np
from scipy.stats import ttest_1samp, wilcoxon, ttest_ind, ranksums


np.random.seed(42)
dataset = np.random.uniform(5, 10, size=(500, 5))


for i in range(5):
    col = dataset[:, i]
    t_statistic, p_value = ttest_1samp(col, 7.5)
    print(f"t-Test on Column {i+1}: t-statistic = {t_statistic}, p-value = {p_value}")


for i in range(5):
    col = dataset[:, i]
    test_statistic, p_value = wilcoxon(col - 7.5)
    print(f"Wilcoxon Signed Rank Test on Column {i+1}: test statistic = {test_statistic}, p-value = {p_value}")


col3 = dataset[:, 2]
col4 = dataset[:, 3]

t_statistic, p_value = ttest_ind(col3, col4)
print(f"Two Sample t-Test on Column 3 and Column 4: t-statistic = {t_statistic}, p-value = {p_value}")

test_statistic, p_value = ranksums(col3, col4)
print(f"Wilcoxon Rank Sum Test on Column 3 and Column 4: test statistic = {test_statistic}, p-value = {p_value}")
```
